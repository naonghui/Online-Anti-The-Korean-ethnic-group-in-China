### 다운로드 받은 폴더를 지정후 파일리스트 생성

```python
file_list = glob.glob(r'/content/drive/MyDrive/riss_test/*.xls')
```

```python
file_list
```

```python
# 다운로드 받은 순서에 맞게 파일리스트 정렬
file_list = natsorted(file_list)
```

### Pandas DataFrame으로 테이블 생성

```python
#빈 데이터프레임 만들기
riss = pd.DataFrame()
```

```python
for file in tqdm(file_list):
    res = pd.read_excel(file,header=None)
    riss = riss.append(res,ignore_index=True)
```

### 통합 테이블 확인

```python
riss
```

### 하나의 엑셀 파일로 저장

절대 경로를 지정하지 않으면, 코랩 임시 저장소에 저장되어서 코랩이 종료되면 사라짐.
