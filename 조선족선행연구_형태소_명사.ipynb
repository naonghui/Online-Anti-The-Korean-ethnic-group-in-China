```python
# 형태소 분석기를 사용하여 특정 품사의 형태소만 수집
# 형태소 분석기별 태그 목록 https://docs.google.com/spreadsheets/d/1OGAjUvalBuX-oZvZ_-9tEfYD2gQe7hTGsgUpiiBSXI8/edit#gid=0

from konlpy.tag import Kkma

kma = Kkma()

test_sent = u"오늘 수업도 재밌으신가요? 아니면 너무 어려우신가요?"  # Unicode로 처리하라는 의미로 u를 붙여 주었습니다.

#꼬꼬마 코드 예제
print('- 꼬꼬마 출력 결과: ')
print("형태소 분할:\t", kma.morphs(test_sent))     # '\t'는 탭입니다. 
print("명사 추출:\t", kma.nouns(test_sent))
print("POS 태깅:\t", kma.pos(test_sent))
print("POS 태깅(flatten):\t", kma.pos(test_sent, flatten=False, join=False))  # 세종 스타일!
print("POS 태깅(join):\t", kma.pos(test_sent, join=True))
print("문장 분할:\t", kma.sentences(test_sent))



```

```python
from konlpy.tag import Komoran   # KOMORAN  @Shineware


# Initiating
kom = Komoran()

test_sent = u"오늘 수업도 재밌으신가요? 아니면 너무 어려우신가요?"  # Unicode로 처리하라는 의미로 u를 붙여 주었습니다.


#코모란 코드 예제
print('- 코모란 출력 결과:') #코모란 코드 예제
print("형태소 분절:\t\t", kom.morphs(test_sent))
print("명사 추출:\t\t", kom.nouns(test_sent))
print("POS 태깅:\t\t", kom.pos(test_sent))
print("POS 태깅(joined):\t", kom.pos(test_sent, flatten=False, join=True))     
```

```python
# 파일을 불러와서 형태소 분석을 해보겠습니다.

with open("/content/조선족선행연구통합.txt", "r") as f:
    text = f.read()
```

```python
from konlpy.tag import Kkma
from collections import Counter 

#명사만 추출
tags = ['NNG','NNP'] #현재 일반명사(nng), 고유명사(nnp), 동사(vv), 형용사(va) 등을 넣을 수 있음.

NN = []
for ej in kma.pos(text): #형태소 분석기 선택
    if ej[1] in tags:
        NN.append(ej[0])


print(NN)

count = Counter(NN)
print(count)
print(count.most_common(10))
```
